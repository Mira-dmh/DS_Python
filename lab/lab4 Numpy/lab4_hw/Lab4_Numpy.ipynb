## Part 3: Aggregations
 

**Problem 11:(4 pts)**<br>
* Create a rank 1 array named `array6` to hold all integers in half-open interval [1,17).
* Reshape it to a rank 2 ndarray of 4 rows and 4 columns. 
* Use **numpy.sum** to find the sum of each column. 
* Use **numpy.sum** to find the sum the whole array. 
* Use **numpy.max()** to find the the largest elements in each row of `array6`. 
* Replace element 4 with a value of 100.
* Use **numpy.argmax()** to find the index of the largest element in `array6`. <br> https://numpy.org/doc/stable/reference/generated/numpy.argmax.html
* Use **numpy.argmax()** to find the index of the largest element in each column of `array6`. 


```python
array6=np.arange(1,17)
print(array6)
array6=array6.reshape(4,4)
print(array6)
print('sum of each column:',array6.sum(axis=0))
print('sum the whole array:',array6.sum())
print('the largest elements in each row of array6:',np.max(array6,axis=1))

filter=array6==4
array6[filter]=100
print('Replace element 4 with a value of 100:\n',array6)

print('the index of the largest element in array6:\n',np.argmax(array6))
print('the index of the largest element in each column of array6:\n',np.argmax(array6,axis=0))
```
