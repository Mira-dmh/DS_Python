### <span style = "color: maroon"> 2. Basic Arithmetic Operations (Element-wise arithmetic operations) </span>

NumPy supports addition (+), subtraction (-), multiplication (*), division (/), exponentiation (), and modulus (%)**.

````python

# Creating NumPy arrays
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

# Element-wise operations
print(f'a + b = {a + b}')   # Addition
print(f'a - b = {a - b}')   # Subtraction
print(f'a * b = {a * b}')   # Multiplication
print(f'a/b={a/b}')
print(f'a / b = {np.round(a / b, 2)}')   # Division
print(f'a ** 2 = {a ** 2}')  # Exponentiation (Square of each element)
print(f'a % 2 = {a % 2}')   # Modulus (Remainder)
````

### <span style = "color: maroon"> 3. Special Numpy Functions for Arithmetic </span>
 NumPy provides optimized functions for element-wise operations.
 | **Operation**   | **Function**               |

| Subtraction   | `np.subtract(a, b)`      |
| Multiplication| `np.multiply(a, b)`      |
| Division      | `np.divide(a, b)`        |
| Power        | `np.power(a, b)`         |
| Modulus       | `np.mod(a, b)`           |
| Round       | `np.round(number, numOfDigits)`           |



````python
print(np.add(a, b))  # Same as a + b
print(np.multiply(a, 2))  # Same as a * 2
````

````python
# importing all the required libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

#source: https://www.analyticsvidhya.com/blog/2021/05/image-processing-using-numpy-with-practical-implementation-and-code/?
````

### 1 \) **Negative of an Image**
Converting a color image to a negative image is very simple. You to perform only 3 steps for each pixel of the image

- First, get the RGB values of the pixel

- Calculate new RGB values using R = 255 ‚Äì R, G = 255 ‚Äì G, B = 255- B

- Finally, save the new RGB values in the pixel

````python
fig = plt.figure(figsize=(10, 10))

fig.add_subplot(1, 2, 1) 
#ÂõæÂΩ¢Á™óÂè£Ë¢´ÂàÜÊàê 1 Ë°å 2 ÂàóÁöÑÂ≠êÂå∫ÂüüÔºå1 Ë°®Á§∫ÈÄâÊã©Á¨¨‰∏Ä‰∏™‰ΩçÁΩÆ„ÄÇ
#Show original image                
plt.imshow(img) #Display the image
plt.title('Original image')

#show negative imge
img_negative = 255 - img #ÂΩ¢ÊàêË¥üÁâá

fig.add_subplot(1, 2, 2)
plt.imshow(img_negative)
plt.title('Negative of RGB image')
````

### 2 \) **Contrast Adjustment**
- We enhance contrast by multiplying pixel values by a scaling factor.
- If img is the original image stored as a NumPy array, we scale it using:

        new_img = img ‚àó factor
 
- The factor should be greater than 1 to increase contrast and less than 1 to reduce contrast.
- We clip values to ensure they remain within the valid pixel range [0, 255].

````python
fig = plt.figure(figsize=(10, 10))

fig.add_subplot(1, 2, 1)
#Show original image                
plt.imshow(img)
plt.title('Original image')

# Apply contrast enhancement (increase contrast)
contrast_factor = 2.5  # Increase contrast by 2.5x 
                        #Note: >1 increases contrast, <1 decreases contrast
img_contrast = np.clip(img * contrast_factor, 0, 255).astype(np.uint8)

#clip ÂáΩÊï∞Â∞ÜÂõæÂÉèÁöÑÂÉèÁ¥†ÂÄºÈôêÂà∂Âú® 0 Âà∞ 255 ‰πãÈó¥ÔºåÂõ†‰∏∫ÂõæÂÉèÁöÑÂÉèÁ¥†ÂÄº‰∏çËÉΩË∂ÖÂá∫Ëøô‰∏™ËåÉÂõ¥„ÄÇ
#Ëã•Êüê‰∏™ÂÉèÁ¥†ÂÄºÂ∞è‰∫é 0Ôºå‰ºöË¢´ËÆæ‰∏∫ 0ÔºõËã•Â§ß‰∫é 255Ôºå‰ºöË¢´ËÆæ‰∏∫ 255„ÄÇ

#.astype(np.uint8)ÔºöÂ∞ÜÂÉèÁ¥†ÂÄºËΩ¨Êç¢‰∏∫ 8 ‰ΩçÊó†Á¨¶Âè∑Êï¥Êï∞Á±ªÂûãÔºå
#Âõ†‰∏∫ÂõæÂÉèÁöÑÂÉèÁ¥†ÂÄºÈÄöÂ∏∏ÊòØ uint8 Á±ªÂûãÔºà0 Âà∞ 255 ‰πãÈó¥Ôºâ„ÄÇ

fig.add_subplot(1, 2, 2)
plt.imshow(img_contrast)
plt.title('Contrast Enhanced Image')
````

### 3 \) **Brightness Adjustment**
- Brightness adjustment formula

        new_img = img + brightness_factor
- Adding a positive value increases brightness.
- Adding a negative value decreases brightness.
- We clip the values to [0, 255] to avoid invalid pixel values.


````python
fig = plt.figure(figsize=(10, 10))

fig.add_subplot(1, 2, 1)
#Show original image                
plt.imshow(img)
plt.title('Original image')

# Convert image to float before modifying brightness
img_float = img.astype(np.float32)  # Prevents overflow, since img.dtype is uint8.
#Â∞ÜÂéüÂßãÂõæÂÉè img ‰ªé uint8 Êï∞ÊçÆÁ±ªÂûãËΩ¨Êç¢‰∏∫ float32„ÄÇ
#ËøôÊ†∑ÂÅöÊòØ‰∏∫‰∫ÜÈÅøÂÖçÂú®Â¢ûÂä†‰∫ÆÂ∫¶Êó∂ÂèëÁîüÊ∫¢Âá∫Ôºå
#Âõ†‰∏∫ uint8 Êï∞ÊçÆÁ±ªÂûãÁöÑÂÉèÁ¥†ÂÄºËåÉÂõ¥ÊòØ 0 Âà∞ 255ÔºåËÄå float32 ÂèØ‰ª•Ë°®Á§∫Êõ¥Â§ßÁöÑÊï∞ÂÄº„ÄÇ

# Increase brightness
brightness_factor = 100  # Increase brightness by adding 50 to each pixel
 
img_bright = np.clip(img_float + brightness_factor, 0, 255).astype(np.uint8)


# Show brightened image
fig.add_subplot(1, 2, 2)
plt.imshow(img_bright)
plt.title('Brightened Image')

plt.show()
````

### **Numpy.astype()**

The .astype() method in NumPy is used to convert the data type (dtype) of an array to a different type.

Syntax: 

        `array.astype(new_dtype)`


- array : The NumPy array to be converted.
- new_dtype : The target data type (np.int32, np.float64, np.uint8, etc.).

````python
arr = np.array([1.9, 2.5, 3.8, 4.2])
arr_int = arr.astype(np.int32)

print(arr_int)       # Output: [1 2 3 4] truncates the decimal part
print(arr_int.dtype) # Output: int32
````

### <span style = "color: maroon"> 4. Aggregation & Statistical Functions </span>

Numpy also provides functions performing summarization across an array, such as computing sums, means, or standard deviations.

| **Function**     | **Description** |
|-----------------|----------------|
| `np.sum(a)`     | Computes the sum of all elements in `a` |
| `np.mean(a)`    | Computes the mean (average) of `a` |
| `np.median(a)`  | Computes the median of `a` |
| `np.std(a)`     | Computes the standard deviation |
| `np.var(a)`     | Computes the variance |
| `np.min(a)`     | Returns the minimum value |
| `np.max(a)`     | Returns the maximum value |



````python
arr2 = np.array([[4, 2, 3],
                [1, 5, 6]])
print(arr2)
````

````python
print(np.sum(arr2))   
print(np.mean(arr2))  
print(np.min(arr2))   
print(np.max(arr2))
````

````python
print(arr2)
#sum up all elements in the ndarray
arr2.sum(axis = 0) #sum up all elements in the columns
````

````python
print(arr2)
#sum up all elements in the ndarray
arr2.sum(axis = 1) #sum up all elements in the columns
````
