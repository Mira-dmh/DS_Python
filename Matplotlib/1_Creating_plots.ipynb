```python
# Load matplotlib.pyplot, pandas, and numpy
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## matplotlib.pyplot.plot
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# Create array for x- and y- coordinates
x = [1, 2, 3, 4]
y = [5, 1, 4, 2]
# Create a 6x5 figure
plt.figure(figsize=[5, 3])
# Create a line plot
plt.plot(x, y)
```

```python
# Create array for x- and y- coordinates
y = [5, 1, 4, 2]
# Create a 6x5 figure
plt.figure(figsize=[5, 3])
# Create a line plot
plt.plot(y, 'ro-')
 # x-axis is automatically generated
```

## np.linspace

```python
plt.figure(figsize=[5, 3])
# Create an equally spaced array with 100
# elements from -5 to 5
x = np.linspace(-5, 5, num=100)

# Plots the graph for y = x^2
plt.plot(x, x * x)

# Plots the graph for y = -4x - 4
plt.plot(x, -4 * x - 4)
```

## Scatter plot for the Old Faithful Geyser Data
- The Eruption column records the duration of each eruption in minutes. It measures how long the geysers eruption lasts.
- The Waiting column records the waiting time to the next eruption in minutes. It measures the duration between the end of one eruption and the beginning of the next.

```python
# Load oldfaithful.csv data
oldFaithful = pd.read_csv('Data_plot/oldfaithful.csv')
oldFaithful.head()
```

```python
# Create a scatter plot
plt.scatter(oldFaithful['Eruption'], oldFaithful['Waiting'])

# Uncomment the following line to save the plot as a png when running locally.
# plt.savefig(fname='oldfaithful1.png')
```
