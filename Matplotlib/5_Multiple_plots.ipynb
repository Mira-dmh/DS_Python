## Multiple plots
`matplotlib.pyplot.subplot`

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html

```python
# Load packages
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load oldfaithfulCluster.csv data
df = pd.read_csv('Data_plot/oldfaithfulCluster.csv')
df.head()
```

```python
plt.figure(figsize = [8, 4])

plt.subplot(2, 1, 1) # 2 rows, 1 column, 1st plot , equivalent to plt.subplot(211)
plt.scatter(df['Eruption'], df['Waiting'])
plt.title('Eruption time vs. waiting time', fontsize=20, c='black')
plt.ylabel('Waiting time', fontsize=14)

plt.subplot(2, 1, 2) # 2 rows, 1 column, 2nd plot, equivalent to plt.subplot(212)
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
plt.scatter(group1['Eruption'], group1['Waiting'], label='1', edgecolors='cyan')
plt.scatter(group0['Eruption'], group0['Waiting'], label='0', edgecolors='red')
plt.xlabel('Eruption time', fontsize=14)
plt.ylabel('Waiting time', fontsize=14)
plt.legend()
```
