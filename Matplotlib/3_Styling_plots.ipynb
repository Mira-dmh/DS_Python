## Styling a plot
`matplotlib.pyplot.plot`

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

````python
# Load packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
# Create array for x- and y- coordinates
x = [1, 2, 3, 4]
y = [1, 5, 3, 4]
# Create a line plot
plt.figure(figsize = [3,2])
plt.plot(x, y, 'b:') # equivalent to: plt.plot(x, y, color='b', linestyle=':')
````

````python
# Create array for x- and y- coordinates
x = [1, 2, 3, 4]
y = [2, 5, 3, 1]
# Create a dot plot
plt.figure(figsize = [3,2])
plt.plot(x, y, 'r^') #red and triangle for data points
````

````python
# Create array for x- and y- coordinates
x = [1, 2, 3, 4]
y = [5, 1, 3, 2]
# Create a dotted-line plot
plt.figure(figsize = [3,2])
plt.plot(x, y, 'gx--') #green, 'x' for data points' and dashed-dot line #equivalent to: plt.plot(x, y, color='g', marker='x', linestyle='--')
````

````python
# Load fatalities.csv data
df2 = pd.read_csv('Data_plot/fatalities.csv')
plt.figure(figsize = [5,3])
# Create line plot for total highway fatalities
plt.plot(df2['Year'], df2['Total'], label='Total', linestyle='--')
# Create line plot for alcohol-related fatalities
plt.plot(df2['Year'], df2['Alcohol-related'], label='Alcohol-related', linestyle=':')

# Create title, font size set to 20
plt.title('Alcohol-related fatalities on highways', fontsize=20)
# Create label for x-axis, font size set to 14
plt.xlabel('Year', fontsize=14)
# Create label for y-axis, font size set to 14
plt.ylabel('Highway fatalities', fontsize=14)
# Create annotation for point (1984, 24762) with text at (1986, 30000)
plt.annotate(
    'Drinking age changed to 21',
    arrowprops=dict(edgecolor='red',linewidth=2,facecolor='black'),
    xy=(1984, 24762),
    xytext=(1986, 30000),
)

# Create legend
plt.legend()
````

## Scatter Plot
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

````python
# Load oldfaithfulCluster.csv data
df = pd.read_csv('Data_plot/oldfaithfulCluster.csv')
# Define groups according to clusters
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
print(group0.head())
print(group1.head())
````

````python

# Create scatter plots
plt.figure(figsize = [5,3])
plt.scatter(group1['Eruption'], group1['Waiting'], 
            label = '1', edgecolors='cyan')
plt.scatter(group0['Eruption'], group0['Waiting'], 
            label='0', edgecolors='red', marker='^', alpha=0.3, s=100) 
#TODO: try marker = '^', 'o', 's', 'd', or 'x'; alpha = 0.5; s = 100

# Add axes labels and title
plt.xlabel('Eruption time', fontsize = 14)
plt.ylabel('Waiting time', fontsize = 14)
plt.title('Eruption time vs. waiting time', fontsize = 20)

# Add legend
plt.legend(title='Cluster')
````

## `*args` and `**kwargs`
- In Python, the use of **\*args** and **\*\*kwargs** as parameters in a function signature allows the function to accept **a variable number of arguments**.
- The **\*args** is used for a variable-length list (**one or more**) of **positional arguments**. 
- The **\*\*kwargs** is used for passing a variable-length list (**zero or more**) of **keyword arguments**.

1. `*args` (Positional Arguments)

`*args` allows passing x-values,  y-values and optional format strings in multiple different ways. For example,

```python
plt.plot([1, 2, 3, 4])  # Only y-values, x-values default to [0, 1, 2, 3]
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])  # x and y values
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], 'ro-')  # x, y, and format string ('r' = red, 'o-' = markers+line)
```

2. `**kwargs` allows passing optional keyword arguments for customization styling, for example,

- color='blue'  Changes line color
- linestyle='dashed'  Changes line style
- marker='o'  Adds circle markers
- linewidth=2  Adjusts line thickness

```python
plt.plot(df2['Year'], df2['Total'],  linestyle='--', marker='o', linewidth=2)
plt.scatter(group0['Eruption'], group0['Waiting'], label = '0', edgecolors='red')
```

````python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure
plt.figure(figsize=(8, 3))

# Generate some data for plotting
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# First subplot setup
plt.subplot(1, 2, 1, facecolor='lightgray')  # First subplot with light gray background
plt.plot(x, y1, label='sin(x)', color='blue')  # Plot first line
plt.plot(x, y2, label='cos(x)', color='green')  # Plot second line
plt.title('First Subplot')
plt.legend(loc = 'lower center')  # Show legend

# Second subplot setup
plt.subplot(1, 2, 2, facecolor='lightgray')  # Second subplot with light gray background
plt.plot(x, y2, label='cos(x)', color='red')  # Plot first line
plt.plot(x, -y1, label='-sin(x)', color='purple')  # Plot second line with negative sine
plt.title('Second Subplot')
plt.legend(loc = 'best')  # Show legend

# Show the plot
plt.show()
````
