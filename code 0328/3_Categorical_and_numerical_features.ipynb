```python
# Import packages
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
tips = sns.load_dataset('tips')
tips.columns = ['Total bill', 'Tip', 'Sex', 'Smoker', 'Day', 'Time', 'Party size']
tips
```

### **One categorical values and one numerical value.**
- `seaborn.stripplot()`
- `seaborn.swarmpplot()`
- `seaborn.kdeplot()`
- `seaborn.histplot()`
- `seaborn.boxplot()`
- `seaborn.violinplot()`
- `seaborn.barplot()`

```python
# Create strip plot for total bill grouped according to day
sns.stripplot(data=tips, x='Total bill', y='Day')
```

```python
# Create strip plot for total bill grouped according to day
plt.figure(figsize=(5,3))
sns.stripplot(data=tips, y='Total bill', x='Day')
```

```python
# Create swarm plot for total bill grouped according to day
plt.figure(figsize=(5,3))
sns.swarmplot(data=tips, x='Day', y='Total bill')
```

```python
# Create kde plot for total bill grouped according to day
plt.figure(figsize=(5,3))
sns.kdeplot(data=tips, x='Total bill', hue='Day')
```

```python
# Create kde plot for total bill grouped according to day
plt.figure(figsize=(9,3))
plt.subplot(121)
sns.histplot(data=tips, x = 'Total bill')

plt.subplot(122)
sns.kdeplot(data=tips, x='Total bill')

plt.title('Histogram and Kernel Density Plot')
```

```python
# Create violin plot for total bill grouped according to day
plt.figure(figsize=(5,3))
sns.boxplot(data=tips, x="Day", y='Total bill')
```

```python
# Create violin plot for total bill grouped according to day
plt.figure(figsize=(5,3))
sns.violinplot(data=tips, x="Day", y='Total bill')
```

### **One categorical values and one numerical value.**
### `seaborn.barplot()` for a bar chart: default Aggregation function: mean

```python
plt.figure(figsize=(4, 3))
sns.barplot(tips, x="Day", y="Tip", errorbar=None )#default estimator='mean'
plt.ylabel("mean Tip")
```


### **Two categorical values and one numerical value.** 

```python
plt.figure(figsize=(5, 3))
sns.barplot(tips, x="Day", y="Tip", hue = 'Sex', errorbar=None )#default estimator='mean'
plt.ylabel("mean Tip")
plt.ylim(0,5)
plt.legend(loc = 'upper left', fontsize = 'small')
```

### `seaborn.barplot()` for a bar chart. Use Aggregation function: np.size
- Create a countplot for one categorical value only. 

```python
import numpy as np
plt.figure(figsize=(4, 3))
#sns.barplot(tips, x="Day", y="Tip", estimator=np.size )#default estimator='mean'
sns.countplot(tips, x="Day") #equivalent to the above.
plt.ylabel("count of Tips")
```

### Note that `seaborn.histplot()` create **gaps** between bars for histograms.

```python
# Create grouped bar chart for total bill grouped according to day
plt.figure(figsize=(5,3))
sns.histplot(data=tips, x='Total bill', hue='Smoker', shrink=0.8, multiple="dodge")
```

```python
# Create stacked bar chart for total bill grouped according to day
plt.figure(figsize=(5,3))
sns.histplot(data=tips, x='Total bill', hue='Smoker', shrink=0.8, multiple="stack")
```
