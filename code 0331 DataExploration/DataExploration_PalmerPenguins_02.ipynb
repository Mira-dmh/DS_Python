```python
# Import packages and dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# Visualize the relationship between numerical features
plt.rcParams.update({'font.size': 14})
p = sns.pairplot(penguins)
plt.show()
```

```python
# Does island affect the relationships?
plt.rcParams.update({'font.size': 14})
p = sns.pairplot(penguins, hue='island', markers=['o', 's', 'v']) #'o':circle, 's': square, 'v': triangle.
plt.show()
```

```python
# Does sex explain the relationships?
plt.rcParams.update({'font.size': 14})
p = sns.pairplot(penguins, hue='sex', markers=['o', 's'])
plt.show()
```

```python
# Does species explain the relationships?
plt.rcParams.update({'font.size': 14})
p = sns.pairplot(penguins, hue='species', markers=['o', 's', 'v'])
plt.show()
```

```python
# Combine sex and species in a single scatterplot
p = sns.scatterplot(
    data=penguins, x='bill_length_mm', y='body_mass_g', hue='species', style='sex'
)
plt.legend(loc=2, prop={'size': 10})
plt.show()
```

### **Contingency table**
A contingency table (also called a cross-tabulation table or cross-tab) is a table used to summarize the relationship between two or more categorical variables by showing how often combinations of their categories occur.

```python
# Contingency table for species and sex
pd.crosstab(index=penguins['species'], columns=penguins['sex'])
```

#### pivot_table() allows more aggregation functions.

```python

penguins.pivot_table(index='species', columns='sex', values='body_mass_g', aggfunc='count')
```

```python
# Grouped bar chart for species and sex
plt.figure(figsize=(5, 4))
p = sns.countplot(data=penguins, x='species', hue='sex', palette='Blues')

plt.legend(loc=1, prop={'size': 10}) #adjust legend location to be outside of the plot #prop={'size': 10} adjusts the font size of the legend
plt.show()
```
