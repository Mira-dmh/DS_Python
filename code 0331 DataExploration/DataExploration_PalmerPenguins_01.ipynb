```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
penguins = pd.read_csv('palmer_penguins.csv')
penguins
```

```python
# Identify the size of the dataset
penguins['species'].unique()
```

```python
# Identify the type and number of present values for each feature
penguins.info()
```

### What Penguin species live in the Torgersen island?

```python
filter = penguins['island'] == 'Torgersen'
penguins[filter].head()
```

```python
# Identify the number of unique species in the filtered dataset
penguins[filter]['species'].unique()
```

### On what Island(s) does the "Gentoo' species live?

```python
filter = penguins['species'] == 'Gentoo'
penguins[filter].head()
```

```python
 
penguins[filter]['island'].unique()
```

### What's the relationship between bill length and body mass?

```python
sns.scatterplot(data=penguins, x = 'bill_length_mm', y = 'body_mass_g')
```

### What's the relationship between bill length and body mass for each species?

```python
sns.scatterplot(data=penguins, x = 'bill_length_mm', y = 'body_mass_g', hue = 'species')
```
